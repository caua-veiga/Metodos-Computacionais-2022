## PDEs parabólicas
### Equação de calor

#### Derivação
Numa haste de metal com temperatura não-uniforme, o calor (energia térmica) é transferido de regiões a temperatura mais elevada para regiões a temperatura mais baixa. Três princípios físicos estão em jogo neste cenário:

1. Calor (ou energia calorífica) de um corpo com propriedades uniformes (homogéneo):

$$\text{Energia calorífica} = c\,m\,u,$$
onde $m$ é a massa do corpo, $u$ é a temperatura, $c$ é o calor especĩfico, com dimensões $[c] =
L^2 T^{-2} \Theta^{-1}$ ($\Theta$ é a dimensão básica de temperatura). $c$ é a energia necessária para elevar uma unidade de massa da substância de uma unidade de temperatura.

2. Lei de Fourier de transferência de calor: a taxa de transferência de calor é proporcional ao simétrico do gradiente de temperatura:

$$\frac{\text{Taxa de transferência de calor}}{\text{área}} = K_0\frac{\partial u}{\partial x} \hspace{30mm} (1)$$
onde $K_0$ é a condutividade térmica, com dimensões $[K_0] = M LT^{-3} \Theta^{-1}$ . Por outras palavras, calor é transferido das regiões de mais alta temperatura para as de mais baixa.

3. Conservação da energia.
Consideremos uma haste uniforme de comprimento $l$ com temperatura não-uniforme, situada no eixo dos $x$, de $x=0$ até $x=l$. Por uniforme queremos dizer que a densidade $\rho$, o calor específico $c$, a conductividade térmica $K_0$ e a secção transversal $A$ são todos constantes ao longo do comprimento da haste.

Suponhamos que os lados da haste estão isoladas e apenas as extremidades podem estar expostas. Suponhamos também que não existe qualquer fonte de calor dentro da haste. 

Consideremos uma fatia muito fina numa posição arbitrária da haste. com espessura $\Delta x$, entre $x$ e $x+\Delta x$. A fatia é tão fina que a temperatura em toda  a sua extensão é igual, $u(x,t)$. Deste modo:

$$
\text{Heat energy of segment} = c \times \rho A \Delta x \times u(x,t) .$$

Por conservação da energia:

$$ \text{mudança de energia térmica da fatia no tempo } \Delta t = \text{calor que entra pela extremidade esquerda}-\text{calor que sai pela extremidade direita}
$$

Destas 3 leis decorre a equaçãpo de calor difusão).
Começando pela lei de Fourier (1),

$$
c \rho A \Delta x u(x,t+\Delta t) -c \rho A \Delta x u(x,t) =
\Delta t A \Big(-K_0 \frac{\partial u}{\partial x} \Big)_x
-\Delta t A \Big(-K_0 \frac{\partial u}{\partial x} \Big)_{x+\Delta x}
$$

Rearranjando os termos (e dado que $\rho, c, A, K_0$ são constantes),

$$\frac{u(x,t+\Delta t) - u(x,t)}{\Delta t} = \frac{K_0}{c \rho} \bigg( \frac{\big(\frac{\partial u}{\partial x} \big)_{x+\Delta x} - \big(\frac{\partial u}{\partial x} \big)_{x}}{\Delta x} \bigg)
$$

Tomando os limites $\Delta x, \Delta t \rightarrow 0$ obtemos a equação de calor:

$$ \frac{\partial u}{\partial t} = \kappa \frac{\partial^2 u}{\partial x^2}
$$

onde $\kappa = \frac{K_0}{c\rho}$ é chamada difusividade térmica, com dimensões $[\kappa]= L^2/T$.

A equação precisa ser complementada com informação adicional, nomeadamente:
 - Condição inicial (CI): nesta situação, a distribuição inicial de temperatura ao longo da haste;
 - Condições fronteira (CF): aqui a temperatura vai depender do que se passa nas extremidades, $x=0,l$. Na realidade, as CF podem ser complicadas. Em geral iremos considerar três casos em cada uma das fronteiras:
  - a temperatura é prescrita na fronteira. Para qualquer $t>0$, $u(0,t)=u_1(t)$. Diz-se de Dirichlet.
  - extremidade isolada: o fluxo de calor é prescrito na fronteira. Para qualquer $t>0$, $-K_0\frac{\partial u}{\partial x}(0,t)=\phi_1(t)$. Diz-se de Neuman.
  - Condições mistas: uma equação envolvendo $u(0,t)$, $\partial u/\partial x(0,t)$, etc. Diz-se de Robin.
  
#### Adimensionalização
  A EDDP tem quantidades físicas com dimensões ($k, l, x, t, u$) e outras podem aparecer nas CI ou CF. Para simplificar a solução e a tornar válida para um leque mais alargado de situações, bem como diminuir a possibilidade de números muito grandes e/ou muito pequenos surjam na computação,  tentamos agrupar tantas quantidades físicas quanto possível. Vamos definir um comprimento, um tempo e uma temperatura característicos $\bar{L}$, $\bar{T}$ e $\bar{\Theta}$, o que nos permite definir variáveis adimensionais:
  
  $$ \hat{x}= \frac{l}{\bar{L}} \quad \hat{t}= \frac{t}{\bar{T}} \quad \hat{u}(\hat{x},\hat{t})= \frac{u(x,t)}{\bar{\Theta}} \quad \hat{f}(\hat{x})= \frac{f(x)}{\bar{\Theta}}.
  $$
  
  A escolha apropriada para $\bar{L}=l$, o comprimento da haste. Deste modo, a variável espacial está limitada a $0<\hat{x} <1$.
  
  A escolha das outras grandezas características é menos óbvia.  Vamos escrever a equação nas novas grandezas. Usando a derivação em cadeia temos:
  
  $$ \begin{align*}
  u_t & = \frac{\partial u}{\partial t} = \bar{\Theta}\frac{\partial \hat{u}}{\partial \hat{t}}\frac{\partial \hat{t}}{\partial t} = \frac{\bar{\Theta}}{\bar{T}}\frac{\partial \hat{u}}{\partial \hat{t}}\\
  u_x & = \frac{\partial u}{\partial x} = \bar{L}\frac{\partial \hat{u}}{\partial \hat{x}}\frac{\partial \hat{x}}{\partial x} = \frac{\bar{\Theta}}{\bar{L}}\frac{\partial \hat{u}}{\partial \hat{x}}\\
  u_{xx} & = \frac{\bar{\Theta}}{\bar{L}^2}\frac{\partial^2 \hat{u}}{\partial \hat{x}^2},
  \end{align*}
  $$
  que substituídos na equação de calor conduz a:
  
  $$ \frac{\partial \hat{u}}{\partial \hat{t}} =  \frac{\bar{T} \kappa}{\bar{L}^2} \frac{\partial^2 \hat{u}}{\partial \hat{x}^2}.
  $$
  
  Para simplificar a equação, escolhemos $\bar{T} = \bar{L}^2/\kappa = l^2/\kappa$, pelo que:
  $$\frac{\partial \hat{u}}{\partial \hat{t}} =   \frac{\partial^2 \hat{u}}{\partial \hat{x}^2}, \qquad 0 < \hat{x} < 1, \quad \hat{t} > 0.
  $$
 
  A escala de tempo (de difusão) característica para o problema é $\bar{T} = l^2/\kappa$.
  
  As CI e CF precisam também ser adimensionalizadas:
  $$ \begin{align*}
  \text{CI:} \quad &  \hat{u}(\hat{x},0)  = \hat{f}(\hat{x}) \quad & 0 < \hat{x} < 1\\
  \text{CF:}\quad &  \hat{u}(0,\hat{t})  = \hat{u}(1,\hat{t})= 0, \quad & \hat{t}>0.
  \end{align*}
  $$
  
#### Derivação alternativa
__Lei de Fourier da difusão__: grandeza flui de regiões de maior para regiões de menor concentração, a uma taxa proporcional ao seu gradiente:
$$ \mathbf{F} = -\alpha \nabla u.$$

Pelo teorema da divergência:

$$\iiint_D \nabla \cdot \mathbf{F} d\,V = \mathrm{fluxo\  através\  de\  S} = \iint(\mathbf{F} \cdot \mathbf{n}) d\,S.
$$


Assim pelo teorema da divergência:
$$ \iiint_D \nabla \cdot \mathbf{F} d\,V = -\frac{\partial}{\partial t}\iiint_D u \ d\,V =  \iiint_D -\frac{\partial u}{\partial t}\  d\,V
$$
verifica-se para qualquer domínio D. Decorre então que 
$$ \nabla \cdot \mathbf{F} = -\frac{\partial u}{\partial t}
$$

Subsituindo  na equação de Fourier vem:
$$ -\frac{\partial u}{\partial t} = \nabla \cdot \mathbf{F} = \nabla \cdot \big(-\alpha \nabla u\big)
$$

Se $\alpha$ não depender do ponto (nem do tempo!) obtemos a equação de difusão, também conhecida por equação de calor, e que tem a forma (normalizada):
$$\frac{\partial u}{\partial t} -
\alpha \frac{\partial^2 u}{\partial x^2}=0,$$
e representa a evolução temporal de uma grandeza sujeita a difusão, isto é, a um processo em que existe uma pressão proporcional ao gradiente da divergência, que leva a uma homogeneização da grandeza no domínio. O segundo membro, aqui nulo, pode conter um termo fonte.  
Como se depreende da presença das derivadas, precisamos de informação adicional para resolver o problema univocamente. A segunda derivada no espaço necessita de 2 condições em todos os pontos da fronteira do domínio. Já a derivada em ordem ao tempo exige o conhecimento da solução em todo o domínio para algum instante de tempo fixo. Em geral será o que chamamos o instante inicial. Note-se que quando o equilíbrio é atingido, a equação tem a forma de uma equação eliptica. As equações de Laplace e Poisson podem assim ser vistas como o estado estacionário da equação de difusão (com e sem fontes, respectivamente).

#### Derivação alternativa, em termos de calor
Consideremos uma região  D (em 1,2 ou 3D, mas também pode ser em $\mathbb{R}^n$). Seja $u(x,t)$ a temperatura num ponto $x=[x_1, x_2,\ldots, x_n]^T$ no instante $t$, e seja $Q(t)$ a quantidade total de calor/energia calorífica contida em D nesse mesmo instante. Seja $c$ o calor específico do material e $\rho$ a sua densidade. Então:
$$ Q(t)=\int_D c \rho u(x,t)\,\text{d}x .$$

A variação de calor em D é assim:
$$ \frac{dH}{dt}= \int_D c \rho u_t(x,t)\,\text{d}x .$$

A lei de Fourier diz que o calor flui das regiões de maior tempaeratura para aquelas de menor temperatura a uma taxa $\kappa>0$ proporcional ao gradiente da temperatura. Por outro lado o calor só pode deixar D pelas fronteiras. Ou seja:
$$ \frac{dH}{dt}= \int_{\partial D} \kappa \nabla u \cdot  n \,\text{d}S ,$$
onde $\partial D$ é a fronteira de D, $n$ é o versor da normal para fora em $\partial D$, e $dS$ é o elemento de área em $\partial D$. Deste modo temos:
$$  \int_D c \rho u_t(x,t)\,\text{d}x = \int_{\partial D} \kappa \nabla u \cdot  n \,\text{d}S.$$

O teorema da divergência diz que:
$$ \int_{\partial D} F \cdot  n \,\text{d}S = \int_D \nabla \cdot F\,\text{d}x,$$
pelo que teremos no nosso caso:
$$ \int_D c \rho u_t(x,t)\,\text{d}x = \int_{D} \nabla \cdot \big(\kappa \nabla u\big) \,\text{d}x.$$

Isto conduz à EDDP:
$$ c \rho u_t = \nabla \cdot \big(\kappa \nabla u\big).$$

No caso particular de um meio isotrópico e homogéneo ($c,\rho$ e $\kappa$ escalares constantes) obtemos a equação de calor:
$$ u_t = \alpha \nabla^2 u ,$$
onde $\alpha=\kappa/c\rho$.


### Problema de condição fronteira inicial para difusão a 1D

Vamos começar por supor condições fronteira de Dirichelet, $u=0$. O problema é então da forma:
$
\begin{align}
\frac{\partial u}{\partial t} &=
\alpha \frac{\partial^2 u}{\partial x^2}, \quad x\in (0,L),\ t\in (0,T]
\tag{1}\\ 
u(x,0) &= I(x), \quad  x\in [0,L]
\tag{2}\\ 
u(0,t) & = 0, \quad  t>0,
\tag{3}\\ 
u(L,t) & = 0, \quad  t>0.
\tag{4}
\end{align}
$

Aqui a função $I(x)$ é conhecida. O parâmetro $\alpha$ é chamado *coeficiente de difusão*.

#### Esquema de Euler para diante (*forward Euler*)
Começamos por discretizar o domínio, tanto no espaço como no tempo, passando do domínio contínuo $[0,L]\times [0,T]$ para o domínio discreto (nodos):

$$\begin{equation*} x_i=i\Delta x,\quad i=0,\ldots,N_x,\end{equation*}$$

e
$$ \begin{equation*} t_n=n\Delta t,\quad n=0,\ldots,N_t .  \end{equation*} $$

Denotamos o valor da função $u$ no ponto de coordenadas $(x_i,t_n)$, $u(x_i,t_n)$, por $u_i^n$, para $i=0,\ldots,N_x$ e $n=0,\ldots,N_t.$  
Substituindo na equação de difusão, em que escolhemos uma discretização para a frente no tempo, e diferenças centrais no espaço, conduz a :
$$ \begin{equation}
\frac{u^{n+1}_i-u^n_i}{\Delta t} = \alpha \frac{u^{n}_{i+1} - 2u^n_i + u^n_{i-1}}{\Delta x^2}.
\tag{7}
\end{equation}$$

Resolvendo em ordem à única quantidade desconhecida, obtemos:

$\begin{equation}
u^{n+1}_i = u^n_i + F\left(
u^{n}_{i+1} - 2u^n_i + u^n_{i-1}\right),
\tag{8}
\end{equation} $

onde $F$ é o número de Fourier da malha:
$\begin{equation}
F = \alpha\frac{\Delta t}{\Delta x^2}.
\tag{9}
\end{equation}$

Este algoritmo pode ser descrito de um modo compacto como:
1. calcular $u^0_i=I(x_i)$ para $i=0,\ldots,N_x$
2. para cada $n=0,1,\ldots,N_t$
   1. aplicar (8) a cada ponto interior  $i=1,\ldots,N_x−1$
   2. aplicar as condições fronteira $ u^{n+1}_i=0$
para $i=0$ e $i=N_x$


O algoritmo pode ser escrito em Python da forma:


```python
import numpy as np
import matplotlib.pylab as plt
%matplotlib inline
```

```python
# Aqui definimos perfis iniciais
def gaussiana(x,L=1.,a = 1., sigma=0.05):

    """perfil gaussiano para condição inicial."""
    return np.exp(-0.5*((x-L/2.0)**2)/sigma**2)

def rect(x,larg=0.1):
    """impulso rectangular para condição inicial."""
    centro = L/2.

    if np.abs(x-centro) > larg:
        return 0
    else:
        return 1
def dois_senos(x, a1=1., k1=1, a2=0.1, k2=100):
    '''2 senos de amplitudes a1 e a2, e frequências 2k1 e 2k2'''
    return a1*np.sin(k1*np.pi*x) + a2*np.sin(k2*np.pi*x)
```

```python
Nx = 64
Nt  = 500
L = 1.
a = 1.
T = 1.1

x    = np.linspace(0, L, Nx+1)    # nodos da grelha no espaço
dx   = x[1] - x[0]
print('dx = {:.5}'.format(dx))
#t    = np.linspace(0, T, Nt+1)    # nodos da grelha no tempo
#dt   = t[1] - t[0]
#F    = a*dt/dx**2
F    = 0.51
dt   = F *dx**2 /a                # calculando dt, devo corrigir número Nt
t    = np.linspace(0, Nt*dt, Nt+1)    # nodos da grelha no tempo
print('número de Fourier = {:.5}'.format(F))
u    = np.zeros(Nx+1)           # incógnita u no novo nível de tempo
u_1  = np.zeros(Nx+1)           # incógnita u no nível de tempo anterior

# Aplicar condição inicial  u(x,0) = I(x)
sigma = 0.05
for i in range(0, Nx+1):
    u_1[i] = gaussiana(x[i],L,a,sigma)
    #u_1[i] = rect(x[i],0.15)
init = u_1.copy()

for n in range(0, Nt):
    # Calcular u em todos os nodos internos
    for i in range(1, Nx):
        u[i] = u_1[i] + F*(u_1[i-1] - 2*u_1[i] + u_1[i+1])
    #u[1:Nx] = u_1[1:Nx] + F*(u_1[0:Nx-1] + 2*u_1[1:Nx] + u_1[2:Nx+1])

    # Aplicar condições fronteira (aqui nem seria necessário! Porquê?)
    u[0] = 0;  u[Nx] = 0

    # Actualizar u_1 antes de passar ao próximo passo
    u_1[:]= u

plt.plot(x,u,'b',x,init,'r--')
plt.show()
```

Experimentemos o que acontece quando a função inicial é uma gaussiana ou uma *cartola*. E experimentemos com diferentes valores de F entre 0 e 1.

Das simulações pode-se concluir que:
 * o esquema de Euler para a frente leva a uma solução divergente se $F>1/2$;
 * a condição inicial em cartola leva a ruído como um dente de serra se $F=1/2$, mas esse ruído não aparece se $F=1/4$,
 * a condição inicial gaussiana leva a uma solução suave se $F=1/2$.

#### Esquema de Euler para trás (*backward  Euler*)

Aplicamos agora uma discretização para trás no tempo, mas novamente diferenças centrais no espaço.
$$ \begin{equation}
\frac{u^{n}_i-u^{n-1}_i}{\Delta t} = \alpha \frac{u^{n}_{i+1} - 2u^n_i + u^n_{i-1}}{\Delta x^2}.
\tag{10}
\end{equation} $$


Supomos que $u^{n-1}_i$ já é conhecido, mas todas as quantidades no "novo" tempo $n$ são desconhecidas. Então já não é possível resolver explicitamente em ordem a  $u^{n}_i$, pois este valor está acoplado aos seus vizinho no espaço, $u^{n}_{i+1}$ e $u^{n}_{i-1}$, que também são desconhecidos nesta fase. O método chama-se por isso *implícito*. Exemplifiquemos com o caso $N_x=3$. A equação (10) escrita para $i=1,\ldots,N_x−1=1,2$ vem:
$$
\begin{align}
\frac{u^{n}_1-u^{n-1}_1}{\Delta t} &= \alpha\frac{u^{n}_{2} - 2u^n_1 + u^n_{0}}{\Delta x^2}
\tag{11}\\ 
\frac{u^{n}_2-u^{n-1}_2}{\Delta t} &= \alpha\frac{u^{n}_{3} - 2u^n_2 + u^n_{1}}{\Delta x^2}
\tag{12}
\end{align}
$$


Os valores na fronteira, $u^{n}_0$ e $ u^{n}_3$, são sabidos (zero no nosso caso, com condições fronteira de  Dirichelet). Separando os novos valores $u^{n}_1$ e $ u^{n}_2$ (incógnitas) no lado esquerdo, conduz a:
$$ \left(\begin{array}{cc}
1+  2F &   - F\\ 
- F    & 1+  2F
\end{array}\right)
\left(\begin{array}{c}
u^{n}_1\\ 
u^{n}_{2}\\ 
\end{array}\right)
=
\left(\begin{array}{c}
u^{n-1}_1\\ 
u^{n-1}_2
\end{array}\right). $$


No caso geral, a eq. (10) dá origem a um sistema de $(N_x -1)$ equações acopladas nas incógnitas $u^{n}_{i}$ para todos os nodos internos $i=1,\ldots,Nx-1$. Isolando as incógnitas do lado esquerdo, teremos:

$$  \begin{equation}
- F u^n_{i-1} + \left(1+  2F \right) u^{n}_i - F u^n_{i+1} =
u_{i-1}^{n-1},
\tag{13}
\end{equation}  $$
para $i=1,\ldots,Nx-1$. 

O sistema de equações (13) pode ser escrito na form matricial:
$\begin{equation*} AU = b\end{equation*},$
onde $U=(u^n_1,\ldots,u^n_{N_x-1})$, e  a matriz tem a estrutura:

$$
\begin{equation}
A =
\left(
\begin{array}{cccccccccc}
A_{1,1} & A_{1,2} & 0 &\cdots & \cdots & \cdots & \cdots & \cdots & 0 \\ 
A_{2,1} & A_{2,2} & 0 & \ddots &   & &  & &  \vdots \\ 
0 & A_{3,2} & A_{3,3} & A_{3,4} & \ddots & &  &  & \vdots \\ 
\vdots & \ddots &  & \ddots & \ddots & 0 &  & & \vdots \\ 
\vdots &  & \ddots & \ddots & \ddots & \ddots & \ddots & & \vdots \\ 
\vdots & &  & 0 & A_{i,i-1} & A_{i,i} & A_{i,i+1} & \ddots & \vdots \\ 
\vdots & & &  & \ddots & \ddots & \ddots &\ddots  & 0 \\ 
\vdots & & & &  &\ddots  & \ddots &\ddots  & A_{N_x-2,N_x-1} \\ 
0 &\cdots & \cdots &\cdots & \cdots & \cdots  & 0 & A_{N_x-2,N_x-1} & A_{N_x-1,N_x-1}
\end{array}  \right)
\tag{14}
\end{equation}
$$

Os (poucos!) elementos não nulos são dados por:

$$\begin{align}
A_{i,i-1} &= -F
\tag{15}\\ 
A_{i,i} &= 1+ 2F
\tag{16}\\ 
A_{i,i+1} &= -F,
\tag{17}
\end{align}$$

enquanto que o lado direito da equação vale:

$$ \begin{equation}
b = \left(\begin{array}{c}
b_1\\ 
b_2\\ 
\vdots\\ 
b_i\\ 
\vdots\\ 
b_{N_x-1}
\end{array}\right)
\tag{18}
\end{equation} $$
com

$$\begin{equation*}
b_i = u^{n-1}_i,\quad i=1,\ldots,N_x-1.
\end{equation*}$$

Note-se que a matriz $A$ tem coeficientes constantes e por isso pode ser calculada uma única vez, no início. Já o vector $b$ precisa ser actualizado em todos os passos temporais.

Um esquema de implementação pode ser:


```python
Nt   = 170
T    = 0.020
Nx = 64
L = 1.
a = 1.
x = np.linspace(0, L, Nx+1)   # # nodos da grelha no espaço
dx = x[1] - x[0]
t = np.linspace(0, T, Nt+1)    #  nodos da grelha no tempo
dt = t[1] - t[0]
u   = np.zeros(Nx+1)          # incógnita u no novo nível de tempo
u_1 = np.zeros(Nx+1)          # incógnita u no nível de tempo anterior

F    = a*dt/dx**2
print('número de Fourier = {:.5}'.format(F))
# Armazenamento para os dados do sistema linear
A = np.zeros((Nx+1, Nx+1))
b = np.zeros(Nx+1)

for i in range(1, Nx):
    A[i,i-1] = -F
    A[i,i+1] = -F
    A[i,i] = 1 + 2*F
A[0,0] = A[Nx,Nx] = 1

# Aplicar condição inicial  u(x,0) = I(x)
for i in range(0, Nx+1):
    #u_1[i] = rect(x[i],0.15)
    u_1[i] = dois_senos(x[i])
init = u_1.copy()

import scipy.linalg

for n in range(0, Nt):
    # Calcular b e resolver o sistema linear
    for i in range(1, Nx):
        b[i] = -u_1[i]
    b[0] = b[Nx] = 0
    u[:] = scipy.linalg.solve(A, b)

    # Actualizar u_1 antes de passar ao próximo passo
    u_1[:] = u

plt.plot(x,u,'b',x,init,'r--')
plt.show()
```

#### Esquema de Crank-Nicholson (CN)

Neste esquema combinamos as diferenças centrais no espaço com uma média dos esquemas precedentes no tempo. Impômos que a equação seja obedecida nos pontos espacias da nossa grelha, mas que no tempo seja obedecida num ponto fictício intermédio, $n+1/2$. O lado direito vem assim:

$\frac{1}{\Delta x^2}\left(u^{n+\frac{1}{2}}_{i-1} - 2u^{n+\frac{1}{2}}_i + u^{n+\frac{1}{2}}_{i+1}\right).$

Mas como calcular $u^{n+\frac{1}{2}}_i$? Uma hipótese é calculá-la como a média dos seus vizinhos no tempo:

$$ u^{n+\frac{1}{2}}_i\approx
\frac{1}{2}\left(u^{n}_i +u^{n+1}_{i}\right).$$

Fazendo o mesmo para todos os pontos espaciais que aparecem no lado direito, calculando a derivada no tempo do lado esquerdo, multiplicando por $\Delta t$, e reunindo termos, vem finalmente:

$$\begin{equation}
u^{n+1}_i - \frac{1}{2} F(u^{n+1}_{i-1} - 2u^{n+1}_i + u^{n+1}_{i+1})
= u^{n}_i + \frac{1}{2} F(u^{n}_{i-1} - 2u^{n}_i + u^{n}_{i+1}).
\tag{19}
\end{equation} $$

Tal como no caso de *backward Euler* também aqui temos equações com incógnitas acopladas, que podem ser escrita na forma matricial tal como nesse caso, com a diferença que as entradas da matriz $A$ são agora dadas por:

$$\begin{align}
A_{i,i-1} &= -\frac{1}{2} F
\tag{20}\\ 
A_{i,i} &= 1 + F
\tag{21}\\ 
A_{i,i+1} &= -\frac{1}{2} F.
\tag{22}
\end{align}$$

O vector do lado direito $b$ tem uma forma semelhante, podendo ser obtido a partir da multiplicação de $U$ por uma matriz tridiagonal, com entradas:

$\begin{align}
B_{i,i-1} &= \frac{1}{2} F
\tag{20}\\ 
B_{i,i} &= 1 - F
\tag{21}\\ 
B_{i,i+1} &= \frac{1}{2} F.
\tag{22}
\end{align}$

#### Esquema da regra $\theta$
Para uma equação genérica:
$$\frac{\partial u}{\partial t} = G[u],\tag{23}$$
em que $G[u]$ é um operador diferencial, a regra $\theta$ escreve-se como:

$$\frac{u^{n+1}_i - u^n_i}{\Delta t} =
\theta G[u^{n+1}_i] + (1-\theta) G[u^{n}_i].\tag{24}$$

A vantagem desta regra é que podemos escrever uma função e escolher um de uma família de esquemas:
* $\theta=0$ dá o esquema de Euler para a frente no tempo (explícito);
* $\theta=1$ dá o esquema de Euler para a trás no tempo (implícito);
* $\theta=\frac{1}{2}$ dá o esquema de Crank-Nicholson no tempo.

#### Uso de matrizes esparsas
A matriz A que encontramos é tridiagonal. Os códigos usados acima usam a totalidade dessa matriz, mesmo sabendo à partida que a maioria dos elementos é zero. Não só armazenamos muitos zeros, como ainda por cima fazemos muitas operações com eles. E numa matriz "inteira" (densa) correspondente a $n_x$ incógnitas, teremos que fazer $\frac{1}{3}n_x^3$ operações e armazenar $n_x^2$ valores. Com algoritmos apropriados é possível diminuir quer a memória para armazenamento, quer o número de operações, a um valor proporcional a $n_x$. O pacote \texttt{scipy.sparse} tem estruturas apropriadas a matrizes esparsas. Em particular pode armazenar diagonais de matrizes. 
O código abaixo mostra como podemos usar essas estruturas para construir a matriz A dos coeficientes nos métodos de "backward Euler" (feito) e Crank-Nicholson.

```python
# Representação de uma matriz esparsa e do vector do lado direito
principal  = np.zeros(Nx+1)
inferior = np.zeros(Nx)
superior = np.zeros(Nx)
b     = np.zeros(Nx+1)

# Precálculo da matriz esparsa
principal[:] = 1 + 2*F
inferior[:] = -F  #1
superior[:] = -F  #1
# Inserir condições fronteira na matriz
principal[0] = 1
principal[Nx] = 1

A = scipy.sparse.diags(
    diagonals=[principal, inferior, superior],
    offsets=[0, -1, 1], shape=(Nx+1, Nx+1),
    format='csr')
print (A.todense())  # Transforma na forma densa da matriz, para verificação.

# Inserir condição inicial
for i in range(0,Nx+1):
    u_1[i] = I(x[i])

for n in range(0, Nt):
    b = u_1
    b[0] = b[-1] = 0.0  # condições fronteira
    u[:] = scipy.sparse.linalg.spsolve(A, b)
    u_1[:] = u
```

### Análise
Como vimos das simulações acima, o comportamento de uma equação de difusão é alargar a forma e diminuir a amplitude de uma distribuição original $u(x,0)=I(x)$.

A equação de difusão admite soluções que dependem de $\eta = (x-c)/\sqrt{4\alpha t}$  para um dado $c$. Uma solução particular é:

$$ u(x,t) = a \text{ erf} (\eta) + b $$

onde:
$$ \text{erf} (\eta) = \frac{2}{\sqrt{\pi}} \int_{0}^{\eta} e^{-\zeta^2}d\zeta $$

é a função erro, e $a$ e $b$ são constantes arbitrárias. A função erro varia entre [-1,1], é ímpar e tende rapidamente para os limites $\pm 1$ ($\text{erf}(3)=0.999977$).

À medida que $t\rightarrow 0$ a função erro tende para uma função degrau centrada em $x=c$. Para um problema de difusão no intervalo [0,1], podemos escolher um degrau em $x=1/2$, para o qual $c=1/2$, $a =-1/2$ e $b=1/2$:

$$u(x,t) = \frac{1}{2} \left(1 - \text{erf}\left( \frac{x-\frac{1}{2}}{\sqrt{4\alpha t}} \right)  \right)
$$

o que implica as condições fronteira:
$$
u(0,t) = \frac{1}{2} \left(1 - \text{erf}\left( \frac{-\frac{1}{2}}{\sqrt{4\alpha t} }\right)  \right)
$$

$$
u(1,t) = \frac{1}{2} \left(1 - \text{erf}\left( \frac{\frac{1}{2}}{\sqrt{4\alpha t} }\right)  \right)
$$

Para $t$ suficientemente pequeno $u(0,t)\approx 1$, $u(1,t)\approx 0$, e quando $t\rightarrow\infty$, $u(x,t)\approx 1/2$ para qualquer $x\in [0,1]$.

#### Solução para um impulso gaussiano 
A equação de difusão admite uma solução da forma:
$$ u(x,t)= \frac{1}{\sqrt{4\pi \alpha t}} \exp \left( -\frac{(x-c)^2}{4\alpha t} \right).
$$

Em $t=0$ é uma função delta de Dirac, que se alarga à medida que o tempo passa.

#### Solução para um seno

$$ u(x,t)=  Q e^{-at}\sin(kx)$$

Escolhendo $Q$ e $k$ decorre da equação que $a= -\alpha k^2$. Isto quer dizer que a forma inicial $I(x) = Q\sin(kx)$ sofre uma atenuação $\exp(-\alpha k^2 t)$, que significa que oscilações rápidas ($k$ grande)  no espaço são atenuadas muito mais rapidamente que oscilações lentas. Isto leva a um suavizar da solução.

Um exemplo é a evolução da condição inicial:
$$ u(x,0) = \sin(\pi x) + 0.1 \sin(100 \pi x) $$

para condições fronteira de Dirichlet no intervalo [0,1].

A solução analítica deste problema é:
$$ u(x,t) = e^{-\pi^2 t}\sin(\pi x) + 0.1 e^{-\pi^2 10^4 t}\sin(100 \pi x) $$

### Análise de equações discretas

A solução acima pode ser escrita numa representação complexa:

$$ u(x,t)=  Q e^{-a t}e^{ikx}.$$

Uma solução geral pode ser escrita à custa de soluções elementares desta forma (componentes de Fourier, na verdade!):
$$ u(x,t) \approx \sum_{k\in K} b_k e^{-\alpha k^2 t}e^{ikx} $$

As amplitudes individuais de cada modo $b_k$ podem ser obtidas da condição inicial:
$$ I(x) \approx \sum_{k\in K} b_k e^{ikx} $$

Podemos aprender bastante do comportamento de métodos numéricos ao investigar como é que uma componente $\exp(−\alpha k^2 t)\exp(ikx)$ é tratada pelo esquema numérico. Tais componentes parecem ser solução dos diferentes esquemas, mas o factor de amplificação varia entre esquemas. Vamos usar a notação $A^n$ para esse factor. O factor exacto é, como vimos, $A_e =  e^{-\alpha k^2 t}$.

#### Análise dos diferentes esquemas
Vamos então procurar soluções da forma:
$$ u_q^n = A^n e^{ikq\Delta x} = A^n e^{ikx}. $$

O factor de amplificação $A$ é determinado inserindo esta escolha no esquema numérico e resolvendo para $A$.

O esquema exacto leva a decaímento, logo devemos requerer que $|A|<1$. No caso de $-1 \le A < 0$, $A^n$  muda de sinal a cada passo temporal, o que indica oscilações (a decair, logo estáveis) não-físicas, que não estão presentes na solução física.

#### Estabilidade
Para determinar quão precisamente um esquema numérico trata uma componente de Fourier, podemos notar que isso corresponde a quão precisamente o factor $A$ aproxima $A_e$. Fazendo um gráfico de $A_e$ e dos diferentes $A$, bem como expandir em série de Taylor a razão $A/A_e$, permite-nos ter uma ideia dessa precisão.

### Análise do esquema de Euler para a frente
Inserindo a componente de Fourier no esquema de Euler para a frente, resolvendo para $A$ e simplificando, vem:
$$ \frac{A-1}{\Delta t} = -\alpha \frac{4}{\Delta x^2}\sin^2\left( \frac{k \Delta x}{2}\right)$$
logo:
$$ A = 1 - 4F\sin^2\left( \frac{k \Delta x}{2}\right)$$
onde, como antes, $F=\alpha \Delta t/\Delta x^2$ é o número de Fourier _numérico_.

A solução numérica completa é assim:
$$ u_q^n = \left( 1 - 4F\sin^2\left( \frac{k \Delta x}{2}\right) \right)^n e^{ikq\Delta x}$$
#### Estabilidade
É evidente que $A<1$. Mas se fôr $A<-1$ o esquema será instável, pois haverá crescimento da componente. Para que $A>-1$ é preciso que:
$$ 4F\sin^2\left( \frac{p}{2}\right) \le 2 $$
($p=k \Delta x$)
ou $$ F \le \frac{1}{2}.$$
ou ainda como condição em $\Delta t$ (uma vez escolhido $\Delta x$):
$$ \Delta t \le \frac{\Delta x^2}{2\alpha}.$$

#### Precisão
Exprimindo $A_e$ também à custa de $p$, temos:
$$ A_e = \exp(-\alpha k^2 \Delta t) = \exp(-4Fp^2).$$

Podemos obter a expansão em série de Taylor do quociente $A/A_e$ em termos do parâmetro $F$:
$$ \frac{A}{A_e} = 1 - 4F \sin^2 p + 2Fp^2 -16F^2p^2\sin^2 p + 8F^2p^4 + \ldots $$
Tendo em conta que $F=\alpha \Delta t/\Delta x^2$, $p=k \Delta x$, e $\sin p <1$ o erro é quando muito:
$$ 1- 4\alpha \frac{\Delta t}{\Delta x^2} + \alpha \Delta t -4 \alpha^2 \Delta t^2 + \alpha^2 \Delta t^2\Delta x^2 \ldots $$

### Análise do esquema de Euler para trás
Inserindo a componente de Fourier no esquema de Euler para trás, resolvendo para $A$ e simplificando, vem:
$$ \frac{1-A^{-1}}{\Delta t} = -\alpha \frac{4}{\Delta x^2}\sin^2\left( \frac{k \Delta x}{2}\right)$$
logo:
$$ A = \left( 1 + 4F\sin^2 p\right)^{-1}$$
onde, como antes, $F=\alpha \Delta t/\Delta x^2$ é o número de Fourier _numérico_.

A solução numérica completa é assim:
$$ u_q^n = \left( 1 + 4F\sin^2 p\right)^n e^{ikq\Delta x}$$
#### Estabilidade
Vê-se que $0<A<1$, ou seja, a solução é estável para qualquer $\Delta t$.


### Análise do esquema de CN
Inserindo a componente de Fourier no esquema de Euler para a frente, resolvendo para $A$ e simplificando, vem:
$$ \frac{A-1}{\Delta t} = -\frac{1}{2}\alpha \frac{4}{\Delta x^2}\sin^2\left( \frac{k \Delta x}{2}\right) (1+A)$$
logo:
$$ A = \frac{1-2F\sin^2 p}{1+2F\sin^2 p}.$$

A solução numérica completa é assim:
$$ u_q^n = \left(\frac{1-2F\sin^2 p}{1+2F\sin^2 p} \right)^n e^{ikq\Delta x}$$
#### Estabilidade
O critério de estabilidade $-1<A<1$ é verificado por qualquer $\Delta t$.
